convolution mathematical technique is used in signal processing. It is composed with multiplication and addition. This convolution technique can be used for identify edges in 2d or 3d images

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signals = np.array([1, 1, 1, 0.5, 0.5, 1, 1, 0.5, 0.5, 0.5])
filter = np.array([-1, 1])
result = np.convolve(signals, filter, 'valid')
```

```python
plt.stem(signals)
plt.show()
```

```python
plt.stem(filter)
plt.show()
```

```python
plt.stem(result)
plt.show()
```

# Image example


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/sample_data/sample.jpg")
img.shape # dimensions of the images : height, width , # of channels
plt.imshow(img)
```

```python
# converting color image to grey scale
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2.shape
plt.imshow(img2, cmap='gray')
```

```python
# extrating vertical edges
v_kernal = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
v_img = cv2.filter2D(img2, -1, v_kernal)
plt.imshow(v_img, cmap='gray')
```

```python
# extracting horizontal edges
h_kernal = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
h_img = cv2.filter2D(img2, -1, h_kernal)
plt.imshow(h_img, cmap='gray')
```
