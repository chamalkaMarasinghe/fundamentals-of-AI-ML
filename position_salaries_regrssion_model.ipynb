```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
salaries = pd.read_csv("/content/sample_data/salaries.csv")
print(salaries)
```

```python
lvl = salaries[["Level"]]
sal = salaries["Salary"]
print(lvl)
print(sal)
```

```python
lin_model = LinearRegression()
lin_model.fit(lvl, sal)
lin_model.score(lvl, sal)
```

```python
poly_model_2 = PolynomialFeatures(degree=2)
lvl_degree_2 = poly_model_2.fit_transform(lvl)
lin_model_2 = LinearRegression()
lin_model_2.fit(lvl_degree_2, sal)
lin_model_2.score(lvl_degree_2,sal)
```

```python
poly_model_3 = PolynomialFeatures(degree=3)
lvl_degree_3 = poly_model_3.fit_transform(lvl)
lin_model_3 = LinearRegression()
lin_model_3.fit(lvl_degree_3, sal)
lin_model_3.score(lvl_degree_3,sal)
```

```python
poly_model_4 = PolynomialFeatures(degree=4)
lvl_degree_4 = poly_model_4.fit_transform(lvl)
lin_model_4 = LinearRegression()
lin_model_4.fit(lvl_degree_4, sal)
lin_model_4.score(lvl_degree_4,sal)
```
