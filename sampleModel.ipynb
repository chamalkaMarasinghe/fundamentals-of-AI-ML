```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
dataFile = pd.read_csv(url)
print(dataFile) # read all rows
```

```python
x = dataFile.drop(["Class"], axis=1) # dataset without column named class
y = dataFile["Class"] # dataset with only column named class
```

```python
from sklearn.model_selection import train_test_split

# we are dividing the whole data set into two categories train and test, train data will be 80% and test data will be 20%(0.2)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(len(y_test))
print(len(x_test))
```

Screenshot (17).png

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout

# building artificial neural network model

model = Sequential()
# hidden layer 1
# units = 10 neurons
# activation function = relu - all negative values coming from upper layers converts to zero
# input dimension size(# of dependent variables)
model.add(Dense(units=10, activation='relu', input_dim=len(x_train.columns)))
# hidden layer 2
# units = 5 neurons
model.add(Dense(units=5, activation='relu'))
# output layer
# units = 1 neurons
# activation function = sigmoid - getting probality value
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

Screenshot (18).png

```python
# compile th emodel
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
# train the model
# epochs - how many times th emodel will be train again and again
# batch size - # of sample data getin at single batch/time
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100, batch_size=32)
```

```python
model.save("weights.h5")
```
